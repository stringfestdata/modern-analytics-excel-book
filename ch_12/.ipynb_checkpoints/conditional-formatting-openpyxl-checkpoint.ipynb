```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Apply conditional formatting to the DataFrame
def highlight_rows(row):
    value = row.loc['body_mass_g']
    if value > 4000:
        color = '#90EE90' # Green
    elif value > 3500:
        color = '#ffffe0' # Yellow
    else:
        color = '' # Blank
    return ['background-color: {}'.format(color) for r in row]
```

```python
formatted_penguins = penguins.style.apply(highlight_rows, axis=1)
formatted_penguins
```

```python
formatted_penguins.to_excel('styled_tips.xlsx', engine='openpyxl',
                                                 index=False)
```
